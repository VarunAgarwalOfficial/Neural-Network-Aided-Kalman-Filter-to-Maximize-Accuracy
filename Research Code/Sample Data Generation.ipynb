```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_len = 100
t = np.linspace(0,100,t_len)
```

```python
a = np.linspace(0,0.75,t_len//6)
a = np.append(a,np.linspace(0 , 0 , t_len//5))
a = np.append(a,np.linspace(-0, -3 , t_len//50))
a = np.append(a,np.linspace(0, 1 , t_len//10))
a = np.append(a,np.linspace(1 , 0 , t_len//10))
a = np.append(a,np.linspace(0 , 0 , t_len//7))
a = np.append(a,np.linspace(0 , -1 , t_len//10))
a = np.append(a,np.linspace(-1, -5 , t_len//50))
a = np.append(a,np.linspace(0, 1 ,t_len-len(a))*0.2)
plt.plot(t,a)
```
